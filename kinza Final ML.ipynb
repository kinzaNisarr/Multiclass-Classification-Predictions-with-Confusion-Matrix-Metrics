# Analyzing and Interpreting Multiclass Classification Predictions with Confusion Matrix Metrics in Python

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support
```

```python
# Sample data
actual_labels = ["Amazon", "Alibaba", "eBay", "Amazon", "Alibaba", "eBay", "Amazon", "Alibaba", "eBay"]
predicted_labels = ["Amazon", "Alibaba", "Alibaba", "Amazon", "eBay", "Alibaba", "eBay", "Amazon", "eBay"]

# Define the class labels
labels = ["Amazon", "Alibaba", "eBay"]

# Compute confusion matrix
cm = confusion_matrix(actual_labels, predicted_labels, labels=labels)
```

```python
# Compute precision, recall, F1 score, and support for each class
precision, recall, f1_score, support = precision_recall_fscore_support(actual_labels, predicted_labels, labels=labels)

# Compute weighted F1 score (average F1 score considering class support)
weighted_f1_score = (f1_score * support).sum() / support.sum()
```

```python
# Create a heatmap
plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

# The interpretation of the confusion matrix:



Amazon:

     2 True Positives (TP): Predicted Amazon and is actually Amazon.
     
     1 False Negative (FN): Predicted not Amazon but is actually Amazon.
     
     1 False Positive (FP): Predicted Amazon but is not actually Amazon.
     
     
Alibaba:

     1 True Positives (TP): Predicted Alibaba and is actually Alibaba.
     
     2 False Negative (FN): Predicted not Alibaba but is actually Alibaba.
     
     2 False Positive (FP): Predicted Alibaba but is not actually Alibaba.
     
     
eBay:

    1 True Positives (TP): Predicted eBay and is actually eBay.
    
    1 False Negative (FN): Predicted not eBay but is actually eBay.
    
    2 False Positive (FP): Predicted eBay but is not actually eBay.

```python
# Print precision, recall, F1 score, and weighted F1 score for each class
for i, label in enumerate(labels):
    print(f"Class: {label}")
    print(f"Precision: {precision[i]}")
    print(f"Recall: {recall[i]}")
    print(f"F1 Score: {f1_score[i]}")
    print()
```

# Interpretation of Class Metrics

```python
# Print weighted F1 score
print(f"Weighted F1 Score: {weighted_f1_score}")
```

# Interpretation of Weighted F1 Score


Weighted F1 Score: 0.44
The weighted F1 score takes into consideration the support (the number of true instances for each label) when averaging the F1 scores, providing a metric that is more representative when dealing with imbalanced datasets.
A score of 0.44 indicates that, on average, the model has moderate predictive power, leaning towards poor performance. Ideally, we want this score to be closer to 1.0 for excellent predictive power.
This score suggests that, while the model has some predictive capabilities, there is substantial room for improvement.
